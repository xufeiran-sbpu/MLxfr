```python
import math 
import matplotlib.pyplot as plt
import numpy as np 

x0 = 10
rho = 2 
u=[]
x=[]
t = np.linspace(-1,10,100000)
for i in range(100000):
    u.append(rho/2*math.exp(-3/2*rho*t[i]))
    x.append(x0*math.exp(-3/2*rho*t[i]))
```

```python
plt.title('x*')    
plt.xlabel('t')
plt.ylabel('x')
plt.plot(t,x,label = 'x0 = 10,rho = 2')
plt.legend()
plt.show()
```

```python
plt.title('u*')    
plt.xlabel('t')
plt.ylabel('u')
plt.plot(t,u,label = 'x0 = 10,rho = 2')
plt.legend()
plt.show()
```
