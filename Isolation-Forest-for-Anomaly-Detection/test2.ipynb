```python
import matplotlib.pyplot as plt
import random
import numpy as np

def fi1(a1,a2):
    return a1**2+4*a2**2
def fi2(a1,a2):
    return (a1-1)**2+(a2-1)**2
N=25
f2=[]
f1=[]
A1=[]
A2=[]
for i in range(N):
    a1 = random.uniform(0,1)
    a2 = random.uniform(0,1)
    f1.append(fi1(a1,a2))
    f2.append(fi2(a1,a2))
    A1.append(a1)
    A2.append(a2)
# a=np.arange(0.01,5,0.01)
# b = a/(4-3*a)
print ("f1:",f1)
print ("f2:",f2)
plt.scatter(f1,f2)
plt.show()
```

```python

dat_1 = np.array(f1).reshape(-1)
dat_2 = np.array(f2).reshape(-1)
data_x = np.vstack((dat_1,dat_2))
a_f = data_x[[0,1]]
a_f = np.array(a_f[:,[1,3,6,8,9,10,12,13,15,18,20,21,23,24]])
plt.scatter(f1,f2,c='blue',label='points')
plt.scatter(a_f[0,:],a_f[1,:],c='r',label='effective points')
plt.xlabel('fi1')
plt.ylabel('fi2')
plt.legend()
plt.show()
```
